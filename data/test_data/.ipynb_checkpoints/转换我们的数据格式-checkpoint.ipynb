```python
def retrieval_by_keywords(d, key_words):
    for i in range(len(key_words) - 1, 1, -1):
        key = "_".join(key_words[:2] + key_words[2:i] + [key_words[-1]])
        if key in d:
            return d[key]
    return []
```

```python
def generate_train_data(retrieval_data, test_data):
    
    one_train_data = {"question":test_data["conversations"][0]["value"], "textbook":test_data["textbook"]}
    
    if test_data["module"] != "教学过程":
        key_words = [
                test_data["grade"],
                test_data["subject"],
                test_data["version"],
                test_data["level"],
                test_data["topic"].replace("@@@@",""),
                test_data["mode"],
                test_data["module"]]
    else:
        key_words = [
                test_data["grade"],
                test_data["subject"],
                test_data["version"],
                test_data["level"],
                test_data["topic"].replace("@@@@",""),
                test_data["module"]]        
    
    retrievaled_lp = retrieval_by_keywords(retrieval_data, key_words)
    
    one_train_data["ctxs"] = retrievaled_lp
    
    return one_train_data, key_words
```

```python
import pandas as pd
import json

def load_prompt(
    prompt_name,
    prompt_path="../prompts/",
):
    task_prompt = open(f"{prompt_path}{prompt_name}.md", "r").read()
    return task_prompt

def generate_question(data, example_source="", teaching_mode_path="../prompts/teach_mode_info.csv"):
    
    teaching_mode_df = pd.read_csv(teaching_mode_path)
    teaching_mode_info = teaching_mode_df[teaching_mode_df["教学模式（新名称）"]==data["tch_mode"]].iloc[0]
    
    if len(data["topic"].strip().split("@@@@")) == 1:
        first_lesson, second_lesson, kcs = data["topic"].split("@@@@")[0], "", ""
    elif len(data["topic"].strip().split("@@@@")) == 2:
        first_lesson, second_lesson, kcs = data["topic"].split("@@@@")[0], data["topic"].split("@@@@")[1], ""
    else:
        first_lesson, second_lesson = data["topic"].split("@@@@")[0], data["topic"].split("@@@@")[1]
        kcs = ",".join(data["topic"].split("@@@@")[2:])
    
    if example_source == "textbooks":
        example = data["textbook"]
    elif example_source == "None" or example_source == "textbooks_new":
        example = ""
    else:
        example = "\n".join(data["ctxs"])
        
    
    if example_source != "textbooks_new":
        task_prompt_system = load_prompt("plan_generator_first_system")
        task_prompt_template = load_prompt("plan_generator_first_template")
        
        task_prompt_template = task_prompt_template.format(
            data["subject"],
            data["module"],
            data["level"],
            data["version"],
            data["grade"],
            first_lesson,
            second_lesson,
            kcs,
            data["tch_mode"],
            teaching_mode_info["模式简介"],
            teaching_mode_info["教学环节"],
            example
        )
    else:
        task_prompt_system = load_prompt("plan_generator_with_textbook_system")
        task_prompt_template = load_prompt("plan_generator_with_textbook_template")
        
        task_prompt_template = task_prompt_template.format(
            data["subject"],
            data["module"],
            data["level"],
            data["version"],
            data["grade"],
            first_lesson,
            second_lesson,
            kcs,
            data["tch_mode"],
            teaching_mode_info["模式简介"],
            teaching_mode_info["教学环节"],
            data["textbook"],
            example
        )
    
    question = f"<Action> generate_module </Action>\n<system start>\n{task_prompt_system}<system end>\n<input start>\n{task_prompt_template}<input end>"
    
    return question
```

```python
all_data = []

for i,row in data.iterrows():
    
    one_item = {"id":row["lp_id"], "module":row["module"]}
    
    questions = generate_question(row, example_source="None")
    
    one_turn = [
        {"from": "human", "value": questions},
        {"from": "gpt", "value": ""},
    ]
    
    one_item["conversations"] = one_turn
    all_data.append(one_item)
```
